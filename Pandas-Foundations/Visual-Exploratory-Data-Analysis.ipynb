```python
# DataFrame üzerinde üç farklı çağrı ile aynı grafik çizdirilebilir
# 1) df.plot(kind='hist')
# 2) df.plt.hist()
# 3) df.hist()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

columns = ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm', 'class']
iris_data = pd.read_csv('iris.csv', index_col=None, names=columns)
print(iris_data.info())
print(iris_data.head())
```

```python
# Default olarak line plot çizilir ve pek açıklayıcı olmadığıgı görülüyor
```

```python
iris_data.plot(x='sepal length in cm', y='sepal width in cm')
plt.show()
```

```python
# Scatter plot daha açıklayıcı
```

```python
iris_data.plot(x='sepal length in cm', y='sepal width in cm', kind='scatter')
plt.show()
```

```python
# Her bir değişkenin ayrı ayrı dağılımlarını incelemek daha bilgilendirici olabilir
```

```python
iris_data.plot(y='sepal length in cm', kind='box')
plt.show()
```

```python
# Histogram ile değişkenin yaklaşık olarak olasılık dağılım fonksiyonu elde edilir
```

```python
iris_data.plot(y='sepal length in cm', kind='hist')
plt.show()
```

```python
# Başka hist argümanları kullanarak histogramları özelleştirebiliriz
# bins parametresi ile kutucukların sayısını belirleriz(ya da aralıkların sayısını)
# range parametresine sağlanan bir tuple ile histogramın alt ve üst sınırları belirlenebilir
# density(normed depreceated oldu) parametresi ile değerlerin 1'e normalize edilip edilmemesi belirlenebilir
# cumulative parametresi ile olasılık yoğunluk fonksiyonunun gösterilmesini sağlayabiliriz
```

```python
iris_data['sepal length in cm'].plot(kind='hist', bins=iris_data.shape[0], normed=True)
plt.show()
```

```python
iris_data['sepal length in cm'].plot(kind='hist', bins=iris_data.shape[0], normed=True, cumulative=True)
plt.show()
```
