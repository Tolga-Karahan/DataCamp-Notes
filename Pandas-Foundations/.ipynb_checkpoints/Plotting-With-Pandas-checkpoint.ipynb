```python
# Veriler okunurken parse_dates=True argümanını sağlayarak tarih sütunlarının datetime64 tipinde olmasını sağlarız
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('AAPL.csv', index_col='Date', parse_dates=True)
df.info()
```

```python
# values metodu ile verileri tutan bir ndarray elde ettikten sora plot fonksiyonuna argüman olarak sağlayarak
# grafik çizebiliriz. plot fonksiyonuna doğrudan Series de verebiliriz böylece biraz daha ayrıntılı grafik elde 
# ederiz. Çünkü series verdiğimiz için plot fonksiyonu indeksleride kullanarak yatay ekseni etiketler
# Doğrudan series üzerinde plot metodunu çağırarak ise daha da detaylı bir grafik elde ederiz. Sütun isimleri de
# çıkartılır
```

```python
close_arr = df.Close.values
close_series = df['Close']

_, plots = plt.subplots(3, 1, figsize=(15, 15))
plots[0].plot(close_arr)
plots[1].plot(close_series)
plots[2] = close_series.plot()
plt.show()
```

```python
# Tüm DataFrame üzerinden de grafik çizdirilebilir. Bu grafikte tüm sütunlar kullanılır
# Aşağıdaki örnekte Volume sütunu diğer hepsini kapattığı için diğer zaman serileri gözükmüyor
```

```python
df.plot()
plt.show()
```

```python
# Diğer grafikleri de görebilmek için logscale kullanılabilir
```

```python
df.plot()
plt.yscale('log')
plt.show()
```

```python
df['Open'].plot(color='r', style='.-', legend=True)
df['Close'].plot(color='b', style='-', legend=True)
plt.show()
```

```python
# Grafikleri kaydetmek için savefig fonksiyonunu kullanıyoruz
```

```python
df['Open'].plot(color='r', style='.-', legend=True)
df['Close'].plot(color='b', style='-', legend=True)
plt.savefig('open-close.pdf')
```

```python
# plot fonksiyonuna subplots=True argümanını sağlarsak her grafik ayrı çizilir
```

```python
df.plot(subplots=True)
plt.show()
```
