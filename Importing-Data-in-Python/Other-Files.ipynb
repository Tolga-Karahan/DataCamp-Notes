```python
# Excel spreadsheet, Matlab dosyaları, SAS dosyaları, Stata dosyaları ve HDF5 dosyaları gibi farklı dosyalardan
# verilerin alınmasına bakacağız. Ayrıca list, dictionary gibi yapıları kaydetmek için pickle kullanırız
# Eğer verilerin okunabilir formatta saklanmasını istiyorsak flat-file olarak kaydedebiliriz. Sadece Python'a
# import edeceksek serialize ederek boyut azaltmak mantıklıdır. pickle paketini serialize etmek için kullanıyoruz
```

```python
import pickle

my_dict={'Ad':'Tolga', 'Soyad:':'Karahan','Yaş':24,'Meslek':'Bilgisayar Mühendisi'}
with open('data.pkl', 'wb') as file:
    pickle.dump(my_dict, file)

with open('data.pkl', 'rb') as file:
    same_dict = pickle.load(file)
    print(same_dict)
```

```python
# Pandas ile Excel spreadsheetlerini de okuyabiliriz. Bu dosyaların içerisinde birden fazla tablo bulunur ve bu
# tabloların isimlerine erişerek istediğimiz tabloyu DataFrame e dönüştürebiliriz
```

```python
import pandas as pd

file='battledeath.xlsx'
xl=pd.ExcelFile(file)
xl.sheet_names
```

```python
# Okuduğumuz excel dosyası üzerinde parse metodunu uygulayarak istediğimiz spreadsheeti dataframe e çevirebiliriz 
```

```python
df=xl.parse('bdonly')
df.head()
```

```python
# SAS dosyalarını import etmek için özel bir context manager olan SAS7BDAT'ı kullanıyoruz. Bunun için ilk olarak
# sas7bdat paketinden SAS7BDAT'ı import ederiz
```

```python
from sas7bdat import SAS7BDAT

with SAS7BDAT('airline.sas7bdat') as file:
    df=file.to_data_frame()

df.head()
```

```python
# stata dosyalarını okumak için farklı bir pakete ihtiyacımız yok. Pandas içerisinde bulunan read_stata metodu
# ile stata dosyalarını okuyabiliriz
```
