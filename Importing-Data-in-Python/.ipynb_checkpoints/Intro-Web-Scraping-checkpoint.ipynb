```python
# BeatifulSoup modülünü kullanarak istek atarak çektiğimiz HTML dosyalarından veri çekebiliriz
# BeatifulSoup modülündeki prettify metodu HTML dosyasını güzel görünecek şekilde yapılandırır
```

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.datacamp.com/teach/documentation'
response = requests.get(url)
html = response.text

soup = BeautifulSoup(html)
print(soup.prettify())
```

```python
# Title, text gibi attributeleri kullanarak etiket içeriklerini çekebiliriz
```

```python
print(soup.title)
#print(soup.text)
```

```python
# find_all metodu ile belirli bir etikete ait tüm içerikleri çekebiliriz
```

```python
a_tags = soup.find_all('a')
for link in a_tags:
    print(link.get('href'))
```
