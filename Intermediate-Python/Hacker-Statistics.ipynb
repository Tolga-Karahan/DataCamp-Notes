```python
# Random Walk olayını implement edelim
```

```python
# Baslangic noktasindayiz
random_walks=[]

# Simülasyonu 10 kez yapalım ve sonuçlara bakalım
for sim in range(10):
    
    # 100 kez zar atıp son ulaştığımız noktayı belirleyeceğiz
    random_walk = [0]
    
    for turn in range(100):
        step = random_walk[-1]

        # Zarı atalım
        dice = np.random.randint(1, 7)

        # Adımımızı hesaplayalım
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1, 7)

        # Hareket edelim
        random_walk.append(step)
     
    random_walks.append(random_walk)

# Simülasyon sonuçlarını görelim
fig, ax = plt.subplots(1, 2, figsize=(10, 5)) 
ax[0].hist(np.array(random_walks)[:,-1])
    

for walk in random_walks:
    ax[1].plot(walk)
plt.show()
```
